```python
categories = []
for key in data_cosmere.keys():
    for cat in data[key]:
        if cat not in categories and cat not in filter:
            categories.append(cat)

for cat in sorted(categories):
    print(cat)
```

```python
for key in data_cosmere.keys():
    data_cosmere[key] = [x for x in data_cosmere[key] if x not in filter]
```

```python
duplicates=[
    ("Denth", "VaraTreledees"),
    ("Vasher", "Zahel"),
    ("Vivenna", "Azure")
]
```

```python
for dup in duplicates:
    data_cosmere[dup[0]] = [x for x in data_cosmere[dup[1]] if x not in data_cosmere[dup[0]]]
    del data_cosmere[dup[1]]
```

```python
keys = [*data_cosmere.keys()]
for key in keys:
    if len(data_cosmere[key]) == 0:
        print("Erasing:", key)
        del data_cosmere[key]
```
