```python
import pandas as pd
import plotnine as pn
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataframe = pd.read_csv("cosmere_gephi_v3.csv", index_col=0)

dataframe
```

```python
print(dataframe[dataframe["Degree"] > 50]["Degree"].count() / dataframe["Degree"].count())

print(dataframe[dataframe["Weighted Degree"] > 50]["Weighted Degree"].count() / dataframe["Weighted Degree"].count())
```

```python
print(dataframe[dataframe["Degree"] > 100]["Degree"].count() / dataframe["Degree"].count())

print(dataframe[dataframe["Weighted Degree"] > 100]["Weighted Degree"].count() / dataframe["Weighted Degree"].count())
```

```python
print(dataframe[dataframe["Degree"] > 200]["Degree"].count() / dataframe["Degree"].count())

print(dataframe[dataframe["Weighted Degree"] > 200]["Weighted Degree"].count() / dataframe["Weighted Degree"].count())
```

```python
graph = pn.ggplot(dataframe, pn.aes(x = "Degree"))
graph += pn.geom_density()
graph += pn.labs(
    title = "Densidad de Grado",
    y = "P(k)", 
    x = "k"
)
graph.save("memoria/img/densidad_grado.png")

print(graph)

graph = pn.ggplot(dataframe, pn.aes(x = "Degree"))
graph += pn.geom_density()
graph += pn.scale_y_log10()
graph += pn.scale_x_log10()
graph += pn.labs(
    title = "Densidad de Grado (Escala logaritmica)",
    y = "P(k)", 
    x = "k"
)
graph.save("memoria/img/densidad_grado_log.png")

print(graph)
```

```python
graph = pn.ggplot(dataframe, pn.aes(x = "Weighted Degree"))
graph += pn.geom_density()
graph += pn.labs(
    title = "Densidad de Grado con Pesos",
    y = "P(k)", 
    x = "k"
)
graph.save("memoria/img/densidad_grado_pesos.png")

print(graph)

graph = pn.ggplot(dataframe, pn.aes(x = "Weighted Degree"))
graph += pn.scale_y_log10()
graph += pn.scale_x_log10()
graph += pn.geom_density()
graph += pn.labs(
    title = "Densidad de Grado con Pesos (Escala logaritmica)",
    y = "P(k)", 
    x = "k"
)
graph.save("memoria/img/densidad_grado_pesos_log.png")

print(graph)
```

```python
graph = pn.ggplot(dataframe, pn.aes(x = "clustering"))
graph += pn.geom_histogram(binwidth=0.05)
graph += pn.labs(
    title = "DistribuciÃ³n del coeficiente de clustering C",
    x = "C de clustering", 
    y = "NÂº de nodos"
)
graph.save("memoria/img/distribucion_clustering.png")

print(graph)
```

```python
graph = pn.ggplot(dataframe, pn.aes(x = "clustering", y="Degree"))
graph += pn.geom_point()
graph += pn.labs(
    title = "Grado k vs coeficiente de clustering C",
    x = "k", 
    y = "C"
)
graph.save("memoria/img/grado_vs_clustering.png")

print(graph)
```
