```python
import json
import numpy as np
import pandas as pd
import copy
```

```python
def search_and_del(catgories, books):
    book = None
    for i, cat in enumerate(catgories):
        if cat in books:
            book = cat
            del catgories[i]
    if book is None:
        book = "Category:Other"
    return book, catgories
```

```python
with open("cosmere.json", "r") as json_file: 
    data = json.load(json_file)
```

```python
data_df_dict = {}
for i, k in enumerate(data.keys()):
    book, new_cat_list = search_and_del(data[k], books)
    book = book.split(":")[1]
    data[k] = new_cat_list
    planet, new_cat_list = search_and_del(data[k], planets)
    planet = planet.split(":")[1]
    data[k] = new_cat_list
    race, new_cat_list = search_and_del(data[k], races)
    race = race.split(":")[1]
    data[k] = new_cat_list
    region, new_cat_list = search_and_del(data[k], regions)
    region = region.split(":")[1]
    data[k] = new_cat_list
    data_df_dict[i] = {"Id": i, "Label": k, "Book" : book, "Planet": planet, "Race": race, "Region": region}
```

```python
connections = np.zeros((len(data), len(data)))
```

```python
unique_categories = []
for k in data.keys():
    for c in data[k]:
        unique_categories.append(c)

unique_categories = [x for x in unique_categories if x not in books]
unique_categories = [x for x in unique_categories if x not in planets]
unique_categories = [x for x in unique_categories if x not in races]
unique_categories = [x for x in unique_categories if x not in regions]

unique_categories = np.unique(unique_categories)
for cat in unique_categories:
    print(cat)
```

```python
data_df = pd.DataFrame.from_dict(data_df_dict).T
```

```python
unique_categories_dict = {}
for cat in unique_categories:
    for i, k in enumerate(data.keys()):
        if cat in data[k]:
            if cat not in unique_categories_dict.keys():
                unique_categories_dict[cat] = [i]
            else:
                unique_categories_dict[cat].append(i)

for k in unique_categories_dict.keys():
    unique_categories_dict[k] = sorted(unique_categories_dict[k])

for k in unique_categories_dict.keys():
    names = ": "
    for v in unique_categories_dict[k]:
        names += data_df.iloc[v]["Label"] + " "
    print(len(unique_categories_dict[k]), ":", k, names)
```

```python
for k in unique_categories_dict.keys():
    if len(unique_categories_dict[k]) > 1:
        for i in range(len(unique_categories_dict[k])):
            for j in range(i+1, len(unique_categories_dict[k])):
                n = unique_categories_dict[k][i]
                m = unique_categories_dict[k][j]
                connections[n][m] += 1
                connections[m][n] += 1
```

```python
edges_pd = {}
count = 0
for i in range(len(connections)):
    for j in range(len(connections[i])):
        if i < j:
            if connections[i][j] != 0.0:
                edges_pd[count] = {"source": i, "target": j, "Type": "Undirected", "Weight": connections[i][j]}
                count+=1
```

```python
clean_index = []

for i in range(len(connections)):
    if np.sum(connections[i]) < 1.0:
        clean_index.append(i)
        print(data_df.iloc[i]["Label"], np.sum(connections[i]))

data_df = data_df.drop(index=clean_index)
```
