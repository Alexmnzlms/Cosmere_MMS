```python
import pandas as pd
import numpy as np
```

```python
dataframes = []

for i in ["0.3", "0.5", "1.0", "1.5", "2.0"]:
    nodes = "modularity/cosmere_gephi_v3_{}.csv".format(i)
    output = "modularity/cosmere_gephi_v3_{}_modularity.csv".format(i)

    nodes_df = pd.read_csv(nodes)

    edges = "cosmere_edges.csv"
    edges_df = pd.read_csv(edges, sep=";")

    total_nodes = len(nodes_df)
    total_edges = len(edges_df)
    print(len(nodes_df), len(edges_df))

    nodes_df = nodes_df[nodes_df["componentnumber"] == 0]
    nodes_filter_ids = nodes_df["Id"].values
    edges_df = edges_df[(edges_df["source"].isin(nodes_filter_ids)) & (edges_df["target"].isin(nodes_filter_ids))]

    print(len(nodes_df), len(edges_df))

    nodes_modularity_clases = nodes_df["modularity_class"]
    nodes_modularity_clases = np.unique(nodes_modularity_clases)

    print(nodes_modularity_clases)

    mod_dict = {}
    for mod_class in nodes_modularity_clases:
        nodes_filter = nodes_df[nodes_df["modularity_class"] == mod_class]
        nodes_filter_ids = nodes_df[nodes_df["modularity_class"] == mod_class]["Id"].values
        edges_filter = edges_df[(edges_df["source"].isin(nodes_filter_ids)) & (edges_df["target"].isin(nodes_filter_ids))]

        mod_dict[mod_class] = {
            "res" : i, 
            "mod_class": mod_class,
            "N nodos": len(nodes_filter),
            "% nodos": len(nodes_filter)/total_nodes,
            "N enlaces": len(edges_filter),
            "% enlaces": len(edges_filter)/total_edges
        }

    dataframe = pd.DataFrame().from_dict(mod_dict).T

    print(dataframe)
    dataframes.append(dataframe)

    dataframe.to_csv(output)

dataframe = pd.concat(dataframes)
dataframe.to_csv("modularity/cosmere_gephi_v3_modularity.csv")
```
